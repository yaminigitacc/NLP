Tokenization: 
-------------
Tokenization is nothing break a complex sentence into words. 
The words in nlp is called as tokens.


Word_Tokenize :
--------------
The process of splitting text into words or tokens. it treats white space and comma(,),stop(.) also as tokens.

Sent_Tokenize :
--------------
Sentenece tokenization is process of break the large set text into small or invididual sentences.

```python
# import library to process text data
import os
import nltk 
import nltk.corpus
```

```python
# Let's take a string try to break it as tokens
AI ='''Artificial Intelligence refers to the intelligence of machines. This is in contrast to the natural intelligence of
humans and animals. With Artificial Intelligence, machines perform functions such as learning, planning, reasoning and
problem-solving. Most noteworthy, Artificial Intelligence is the simulation of human intelligence by machines.
It is probably the fastest-growing development in the World of technology and innovation. Furthermore, many experts believe
AI could solve major challenges and crisis situations.'''
```

```python
from nltk.tokenize import word_tokenize
word_token = word_tokenize(AI)
word_token
```

```python
# check type of AI
type(AI)
```

```python
len(word_token)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
# sent_tokenize will break the large text data into small individual sentences.
AI_sent = sent_tokenize(AI)
AI_sent
```

```python
len(AI_sent)
```

# Blankline_tokenize
-- How many paragraphs in input text

```python
# blankline_tokenize
from nltk.tokenize import blankline_tokenize

AI_blt = blankline_tokenize(AI) # gives us how many paragraphs in our input
AI_blt
```

```python
len(AI_blt)
```

# Whitespace_Tokenizer
--he text is split wherever there is a whitespace character.

```python
# whitespace_tokenizer 
# The text is split wherever there is a whitespace character.
# It doesn't account for punctuation, which means punctuation marks will be considered part of the adjacent word.
# when we compared to word_tokenizer with whitespace the whitespace will not treate comma or (.) as a token where as it is in word_tokenizer.

from nltk.tokenize import WhitespaceTokenizer
wt = WhitespaceTokenizer().tokenize(AI)
wt
```

```python
len(wt)
```

# Wordprunct_Tokenize
-- we did not call numbers or any special chars to be as tokens in such cases to allow them treat as tokens we need use wordpunct_tokens


```python
s ="good apple cost is $3.88 in hyderabad,please buy two of them.Thanks"
s
```

```python
#wordpunct_tokenize
# If we observe there is apple cost $3.88 ,
#usually we did not call numbers or any special chars to be as tokens in such cases to allow them treat as tokens we need use wordpunct_tokens
from nltk.tokenize import wordpunct_tokenize
wpt = wordpunct_tokenize(s)
wpt
```

```python
# if we see AI tokens has incresed with wordpunct_tokenize.
w_p = wordpunct_tokenize(AI)
w_p
len(w_p)
```

# Stemming:
 ---Normalises the words into root form or base form

```python
from nltk.stem import PorterStemmer
pst = PorterStemmer() # 
```

```python
pst.stem('affection')
```

```python
pst.stem('pefection')
```

```python
pst.stem('playing')
```

```python
pst.stem('maximum')
```

```python

words_to_stem = ['give','giving','given','gave']

for words in words_to_stem:
    print(words+":" +pst.stem(words))
```

```python
word_to_stem =['give','giving','given','gaved','thinking','loving','maximum']
for words in word_to_stem:
     print(words+":" +pst.stem(words))
```

```python
# lets apply lancasterstemmer to words
# lancasterStemmer gives us core root form words
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()

for words in word_to_stem:
     print(words+":" +lst.stem(words))
```

```python
# for which language we need to stem to be mentioned for snowball stemmer'
# snowball stemmer act like as porter stemmer
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')

for words in word_to_stem:
    print(words+':'+sbst.stem(words))
```

# NER : Named Entity Recognition

```python
from nltk import ne_chunk
```

```python
NE_sent = 'The US president stays in the WHITEHOUSE'
```

```python
NE_tokens = word_tokenize(NE_sent)
NE_tokens
```

```python
NE_tags = nltk.pos_tag(NE_tokens)
NE_tags
```

```python
NE_NER = ne_chunk(NE_tags)
print(NE_NER)
```
